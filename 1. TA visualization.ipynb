### Create average grade column (AvgGrade) column based on G1,G2 & G3

```python
# get average grade from G1, G2 and G3
df['AvgGrade'] = (df['G1'] + df['G2'] + df['G3']) / 3

df['AvgGrade'].describe()
```

## Creating Grade and Pass or Fail column based on average values

create column 'grading' based on Average Grade:

| Letter| Grade (%)| Total (out of 20)|
|------|------|------|
|A|	  90%-100%|	18-20|
|B|	  80%-89%|	16-17.9|   
|C|	  70%-79%|	14-15.9|
|D|	  60%-69%|	12-13.9|
|F|	  < 59%|<11.9|


```python
def define_grade(df):
    # Create a list to store the df
    grades = []
    # For each row in the column,
    for row in df['AvgGrade']:
        # if more than a value,
        if row >= 18: 
            # Append a letter grade
            grades.append('A')
        # else, if more than a value,
        elif row >= 16:
            # Append a letter grade
            grades.append('B')
        # else, if more than a value,
        elif row >= 14: 
            # Append a letter grade
            grades.append('C')   
        elif row >= 12: 
            # Append a letter grade
            grades.append('D') 
        elif row <= 11.9:
            # Append a letter grade
            grades.append('F') 
    # Create a column from the list
    df['grades'] = grades
    return df
df = define_grade(df)
print(df['grades'].value_counts())

# Plot Grades vs No of Students
plt.figure(figsize=(10,6))
ax = sns.countplot(x='grades',data=df,order=df['grades'].value_counts().index);
ax.set_xticklabels(ax.get_xticklabels(), fontsize=8);
plt.title("Grades vs no. of Students")
```

```python
# Create Pass or Fail Column based on average value
def define_PorF(df):
    PorFs = []
    # For each row in the column,
    for row in df['AvgGrade']:
        # if more than a value,
        if row >= 12: 
            # Append a letter PorF
            PorFs.append('P')
        elif row <= 11.9:
            # Append a letter PorF
            PorFs.append('F') 
    # Create a column from the list
    df['PorF'] = PorFs
    return df
df = define_PorF(df)
# print(df['PorF'].value_counts())
```
